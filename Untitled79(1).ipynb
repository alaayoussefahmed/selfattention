```python
import numpy as np

# Sentence
sentence = ["What", "are", "the", "symptoms", "of", "diabetes"]

# Assign random embeddings (6 tokens, embedding size = 8)
np.random.seed(0)
X = np.random.rand(len(sentence), 8)

# Weight matrices
WQ = np.random.rand(8, 8)
WK = np.random.rand(8, 8)
WV = np.random.rand(8, 8)

# Compute Q, K, V
Q = np.dot(X, WQ)
K = np.dot(X, WK)
V = np.dot(X, WV)

# Scaled Dot-Product Attention
dk = K.shape[1]
scores = np.dot(Q, K.T) / np.sqrt(dk)

# Softmax function
def softmax(x):
    return np.exp(x) / np.sum(np.exp(x), axis=-1, keepdims=True)

attention_weights = softmax(scores)

# Output
output = np.dot(attention_weights, V)

# Print results
print("Tokens:", sentence)
print("\nAttention Weights:\n", attention_weights)
print("\nOutput:\n", output)
```
